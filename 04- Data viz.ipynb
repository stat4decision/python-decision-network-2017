```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
plt.plot(np.random.randn(100),label="action 1")
plt.plot(np.random.randn(100),label="action 2")
plt.plot()
plt.title("Mon graphique")
plt.ylabel("Actions")
plt.xlabel("Temps")
plt.legend()
```

```python
plt.hist(np.random.randn(100000),bins=40,label="action 1",alpha=0.5)
plt.hist(np.random.randn(100000),bins=40,label="action 2",alpha=0.5)

plt.title("Mon graphique")
plt.ylabel("Actions")
plt.xlabel("Temps")
plt.legend()
```
