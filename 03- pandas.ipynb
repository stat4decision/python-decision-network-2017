```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

Définir un objet Series comprenant la population de 5 pays puis afficher les pays ayant une population > 50’000’000.


```python
ser1=Series([70,40,300,8,1300],index=["France", "Espagne", "USA", "Suisse",
                                      "Chine"])
```

```python
ser2=ser1[ser1>50]
```

```python
#on ajoute deux séries en remplaçant les données manquantes par 0
ser3=ser1.add(ser2,fill_value=0)
```

```python
#on ajoute les noms de la série et des index
ser3.name="Population"
ser3.index.name="Pays"
ser3
```

```python
#on importe du SQL
from sqlalchemy import create_engine
```

```python
engine=create_engine('sqlite:///Salaries.sqlite')
```

```python
engine.table_names()
```

```python
frame_salaries=pd.read_sql_query('Select * from Salaries', engine)
```

```python
frame_salaries.shape
```

```python
frame_salaries.info()
```

```python
#on affiche la moyenne des salaires
frame_salaries["TotalPay"].mean()
```

```python
frame_salaries["TotalPay"].describe()
```

```python
frame_salaries.sort_values("TotalPay",ascending=False)
```

```python
#on affiche les différents métiers que l'on trie
frame_salaries["JobTitle"].value_counts().sort_values()
```

```python
frame_salaries["JobTitle"].unique()
```

```python
#pour préparer les données, on passe tous les jobs en majuscules
frame_salaries["JOB"]=frame_salaries["JobTitle"].str.upper()
```

```python
frame_salaries["JOB"].nunique()
```

```python
#discretisation intervalles égaux
pd.cut(frame_salaries["TotalPay"],5).value_counts()
```

```python
#discretisation fréquences égales
frame_salaries["TotalPay Disc"]=pd.qcut(frame_salaries["TotalPay"],5)
```

```python
frame_salaries
```

```python
#on crée une structure groupby permettant d'étudier chaque JOB
group_salaries=frame_salaries.groupby("JOB")
```

```python
frame_mean=group_salaries.mean()
```
